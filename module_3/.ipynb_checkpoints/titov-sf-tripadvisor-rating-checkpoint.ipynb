### 1.3 Cuisine Style: Кухня

```python
# Cuisine Style - суммарно 11590 пропусков, сохраним по ним информацию в отдельной переменной
df['NAN_Cuisine Style'] = pd.isna(df['Cuisine Style']).astype('uint8')

# заполним пропуски значением 'Other'
df['Cuisine Style'] = df['Cuisine Style'].fillna("['Other']")

# проведем обработку значений переменной, превратив строку в список
df['Cuisine Style'] = df['Cuisine Style'].apply(
    lambda x: x[1:-1].replace("\'", '').split(', '))
```

```python
# добавим новый признак, отобращающий количество кухонь в ресторане
df['Cuisine_counts'] = df['Cuisine Style'].apply(lambda x: len(x))
```

```python
# посмотрим распределение получившегося признака
df['Cuisine_counts'].hist(figsize=(7, 4), bins=100)
```

```python
# посмотрим солько кухонь участвует в датасете
# для этого вынесем весь перечень уникальных кухонь в 1 переменную
cuisines_set = set()
for lst in df['Cuisine Style'].tolist():
    for cuisine in lst:
        cuisines_set.add(cuisine)

len(cuisines_set)
```

```python
# необходимо отнормировать критерий Ranking по городам City
City_mean_Ranking = df.groupby(['City'])['Ranking'].mean()
City_count_Restorant = df['City'].value_counts(ascending=False)
df['City_mean_Ranking'] = df['City'].apply(
    lambda x: City_mean_Ranking[x])
df['City_count_Restorant'] = df['City'].apply(
    lambda x: City_count_Restorant[x])
df['City_norm_Ranking'] = (
    df['Ranking'] - df['City_mean_Ranking']) / df['City_count_Restorant']
```

```python
# снова посмотрим на топ 10 городов
for x in (df['City'].value_counts())[0:10].index:
    df['City_norm_Ranking'][df['City'] == x].hist(bins=100)
plt.show()
```

### 1.5 Price Range: Цены в ресторане в 3 категориях

```python
# Price Range - суммарно 17361 пропусков, сохраним информацию в отдельную переменную
df['NaN_Price Range'] = pd.isna(df['Price Range']).astype('uint8')

# обработаем значения в столбце
dict_Price_Range = {'$': 1, '$$ - $$$': 2, '$$$$': 3}
df['Price Range'] = df['Price Range'].map(lambda x: dict_Price_Range.get(x, x))

# заполняем значения 2кой, т.к. большинство ресторанов по заполненным данным имеют средний разброс цен ($$ - $$$)
df['Price Range'] = df['Price Range'].fillna(2)
```

### 1.6 Number of Reviews: Количество отзывов    

```python
# Number of Reviews - 3200 пропусков
df['NAN_Number_of_Reviews'] = pd.isna(df['Number of Reviews']).astype('uint8')

# Заполним пропуски 0
df['Number of Reviews'] = df['Number of Reviews'].fillna(0)
```

```python
# посмотрим распределение
df['Number of Reviews'].hist()
```

### 1.7 Reviews: 2 последних отзыва и даты этих отзывов

```python
# Review - 2 пропуска в тестовом датасете, заполним их
df['Reviews'] = df['Reviews'].fillna('[[], []]')

# 8114 строк со значением [[], []], что тоже является пустыми значениями, сохраним их в отдельную переменную
df['NAN_Reviews'] = (df['Reviews'] == '[[], []]').astype('uint8')
```

```python
# вытащим даты из данного признака в отдельный столбец
pattern = re.compile('\d\d\/\d\d\/\d\d\d\d')
df['reviews_date'] = df.Reviews.apply(lambda x: pattern.findall(x))
```

```python
# добавим новый признак с количеством дат
df['len_date'] = df['reviews_date'].apply(lambda x: len(x))
```

```python
# посмотрим какие получились значения
df['len_date'].value_counts()
```

```python
df['reviews_date'] = df.reviews_date.apply(
    lambda x: None if len(x) == 0 else x)
```

```python
# вытягиваем дату 1го отзыва
df['review_date_1'] = pd.to_datetime(
    df.reviews_date.apply(lambda x: x[0] if x != None else x))
```

```python
# вытягиваем дату 2го отзыва
df['review_date_2'] = pd.to_datetime(
    df.reviews_date.apply(lambda x: x if x == None else(x[1] if len(x) == 2 else x[0])))
```

```python
df['review_date_1'].hist(figsize=(10, 5), bins=20)
df['review_date_2'].hist(figsize=(10, 5), bins=20)
plt.tight_layout()
```

Даты практически идентичные

```python
# добавим новый признак - разница между 1 и 2 отзывом в днях
df['reviews_data_appart'] = (
    abs(df['review_date_1'] - df['review_date_2']).dt.days.fillna(0))
```

```python
df['reviews_data_appart'].hist(figsize=(10, 5), bins=20)
plt.tight_layout()
```

```python
# Теперь выделим тестовую часть
train_data = df.query('sample == 1').drop(['sample'], axis=1)
test_data = df.query('sample == 0').drop(['sample'], axis=1)

y = train_data.Rating.values            # наш таргет
X = train_data.drop(['Rating'], axis=1)
```

**Перед тем как отправлять наши данные на обучение, разделим данные на еще один тест и трейн, для валидации.

```python
# Воспользуемся специальной функцие train_test_split для разбивки тестовых данных
# выделим 20% данных на валидацию (параметр test_size)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_SEED)
```

```python
# проверяем
test_data.shape, train_data.shape, X.shape, X_train.shape, X_test.shape
```

# 3. Model 
Сам ML

```python
# Импортируем необходимые библиотеки:
# инструмент для создания и обучения модели
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics  # инструменты для оценки точности модели
```

```python
# Создаём модель (НАСТРОЙКИ НЕ ТРОГАЕМ)
model = RandomForestRegressor(
    n_estimators=100, verbose=1, n_jobs=-1, random_state=RANDOM_SEED)
```

```python
# Обучаем модель на тестовом наборе данных
model.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = model.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```

```python
# в RandomForestRegressor есть возможность вывести самые важные признаки для модели
plt.rcParams['figure.figsize'] = (10, 10)
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(15).plot(kind='barh')
```
